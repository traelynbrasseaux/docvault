# Chunking Strategy Exploration

Visual comparison of the three chunking strategies available in docvault:
- **Fixed-size**: splits by character count with configurable overlap
- **Recursive character**: splits by paragraph → sentence → word boundaries
- **Semantic**: merges adjacent sentences by embedding cosine similarity

```python
import sys
sys.path.insert(0, '../src')

from docvault.config import Config
from docvault.ingest.chunking import chunk_text, ChunkingStrategy
from docvault.embeddings.encoder import Encoder
```

```python
# Sample text for comparison
sample_text = """
Retrieval-Augmented Generation (RAG) is a technique that combines information retrieval 
with text generation. It was introduced by Lewis et al. in 2020 and has since become 
a foundational approach in modern NLP systems.

The core idea is straightforward. Given a user query, the system first retrieves relevant 
documents from a knowledge base. These retrieved passages are then provided as context 
to a language model, which generates a grounded response.

This approach has several advantages over pure parametric models. First, it allows the 
model to access up-to-date information without retraining. Second, it provides 
interpretability through source citations. Third, it reduces hallucination by anchoring 
responses to retrieved evidence.
"""

cfg = Config()
print(f"Text length: {len(sample_text)} chars")
```

```python
# --- Fixed-size chunking ---
fixed_chunks = chunk_text(
    text=sample_text,
    doc_name='sample',
    page_number=1,
    strategy=ChunkingStrategy.FIXED,
    chunk_size=200,
    chunk_overlap=40,
)
print(f"Fixed-size: {len(fixed_chunks)} chunks")
for i, c in enumerate(fixed_chunks):
    print(f"  [{i}] {c.text[:80]!r}...  ({c.metadata.char_count} chars)")
```

```python
# --- Recursive character chunking ---
recursive_chunks = chunk_text(
    text=sample_text,
    doc_name='sample',
    page_number=1,
    strategy=ChunkingStrategy.RECURSIVE,
    chunk_size=200,
    chunk_overlap=40,
)
print(f"Recursive: {len(recursive_chunks)} chunks")
for i, c in enumerate(recursive_chunks):
    print(f"  [{i}] {c.text[:80]!r}...  ({c.metadata.char_count} chars)")
```

```python
# --- Semantic chunking ---
encoder = Encoder(cfg)
semantic_chunks = chunk_text(
    text=sample_text,
    doc_name='sample',
    page_number=1,
    strategy=ChunkingStrategy.SEMANTIC,
    chunk_size=200,
    chunk_overlap=40,
    encoder=encoder,
)
print(f"Semantic: {len(semantic_chunks)} chunks")
for i, c in enumerate(semantic_chunks):
    print(f"  [{i}] {c.text[:80]!r}...  ({c.metadata.char_count} chars)")
```
